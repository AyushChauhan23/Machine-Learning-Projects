```python
plt.figure(figsize=(20,10))
tc=df.corr().round(2)
sns.heatmap(tc,annot=True)
```

```python
target=tc["medv"].sort_values(ascending=False)
top_positive_correlations=target.head()
top_negative_correlations=target.tail()
print('positive correlations:\n',top_positive_correlations)
print('negative correlations:\n',top_negative_correlations)
```

```python
x=df.drop('medv',axis=1)
y=df['medv']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
print('coefficient:',model.coef_)
print('intercept:',model.intercept_)
```

```python
y_pred=model.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
print(mae)
print(mse)
print(rmse)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(y_test,y_pred,color='green')
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()],color='red',linewidth=2)
plt.title('Actual VS Predicted House Prices')
plt.ylabel('Predicted House Prices')
plt.xlabel('Actual House Prices')
plt.show()
```
